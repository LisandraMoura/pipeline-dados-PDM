```python
# PreparaÃ§Ã£o: adicionar uma coluna de renda mÃ©dia (V005 + V009)
df['RENDA'] = df['V005'] + df['V009']
```

### InformaÃ§Ãµes sobre os dados relevante
- nÃºmero do partido, quando voto em legenda; 

- nÃºmero 95, quando voto em branco; 

- nÃºmero 96, quando voto nulo; 

- nÃºmero 97, quando voto anulado e apurado em 
separado.

```python
# 1. NR_VOTAVEL com maior nÃºmero de votos por Nome_do_subdistrito
nr_votavel_subdistrito = df.groupby(['Nome_do_subdistrito', 'NR_VOTAVEL'])['QT_VOTOS'].sum().reset_index()
nr_votavel_max = nr_votavel_subdistrito.loc[nr_votavel_subdistrito.groupby('Nome_do_subdistrito')['QT_VOTOS'].idxmax()]

# Partidos polÃ­ticos mais bem votados por bairro
nr_votavel_max.sort_values('QT_VOTOS')
```

```python
# 2. NM_VOTAVEL com maior nÃºmero de votos por Nome_do_subdistrito
# Candidato com maior numero de votos por bairro

nm_votavel_subdistrito = df.groupby(['Nome_do_subdistrito', 'NM_VOTAVEL'])['QT_VOTOS'].sum().reset_index()
nm_votavel_max = nm_votavel_subdistrito.loc[nm_votavel_subdistrito.groupby('Nome_do_subdistrito')['QT_VOTOS'].idxmax()]
nm_votavel_max.sort_values('QT_VOTOS')
```

```python
# 3. NM_VOTAVEL com maior nÃºmero de votos por renda
# Ordenar por renda para pegar as rendas mais altas e baixas
nm_votavel_renda = df.groupby(['RENDA','NR_VOTAVEL','NM_VOTAVEL'])['QT_VOTOS'].sum().reset_index()
renda_sorted = nm_votavel_renda.sort_values(by='RENDA')

# Candidato e partido com maior quantidade de votos por renda (renda baixa e alta)
renda_sorted
```

```python
# RENDA MAIS BAIXOS
renda_baixa = renda_sorted.head(5)
renda_baixa
```

```python
# RENDAS MAIS ALTAS
renda_alta = renda_sorted.tail(5)
renda_alta
```
