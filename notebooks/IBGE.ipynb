### transformando os dados

Tratamento em strings
Tratamento em nulos (como as variaveis nulas foram todas sobre renda, entendemos como inexistente. Além disso, o valor de nulos diante do dataset é praticamente irrelevante)

```python
ibge.shape
```

```python
# Colunas com valores nulos
print(ibge.isnull().sum())
```

```python
from unidecode import unidecode

# transformações nos dados
# 
# Função para transformar texto: remover acentos e deixar em maiúsculas
def transformar_string(valor):
    if isinstance(valor, str):  # Verifica se é uma string
        return unidecode(valor).upper()
    return valor  # Retorna o valor original se não for string

# Aplica a transformação em todas as colunas de tipo string
for col in ibge.select_dtypes(include=["object"]).columns:
    ibge[col] = ibge[col].apply(transformar_string)
```

```python
# Substituir todos os valores nulos no DataFrame por zero
ibge.fillna(0, inplace=True)

# Verificar se ainda existem valores nulos
print(ibge.isnull().sum())  # Deve mostrar zero para todas as colunas
```
