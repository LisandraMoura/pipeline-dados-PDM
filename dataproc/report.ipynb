```python
import pandas as pd 

df = pd.read_csv('gs://trabalho-final-pdm/gold/tabelas_knn.csv', storage_options={"token": "google_default"})
df
```

```python
# Quantidade de votos por partido por ano
votos_por_partido_ano = df.groupby(['ANO_ELEICAO', 'PARTIDO'])['QT_VOTOS'].sum().reset_index()

# Filtrando para o ano de 2016 e ordenando
votos_2016 = votos_por_partido_ano[votos_por_partido_ano['ANO_ELEICAO'] == 2016].sort_values('QT_VOTOS')

print(votos_2016)
```

```python
# Quantidade de mulheres que votaram nos partidos por ano
# Multiplicando votos pela proporção de mulheres
df['Votos_Feminino'] = df['QT_VOTOS'] * df['Proporcao_feminino']
mulheres_por_partido_ano = df.groupby(['ANO_ELEICAO', 'PARTIDO'])['Votos_Feminino'].sum().reset_index()

# Filtrando para o ano de 2016 e ordenando
x = mulheres_por_partido_ano[mulheres_por_partido_ano['ANO_ELEICAO'] == 2020].sort_values('Votos_Feminino')

x
```

```python
# Quantidade de negros que votaram nos partidos por ano
# Multiplicando votos pela proporção de negros
df['Votos_Negros'] = df['QT_VOTOS'] * df['Proporcao_negros_pardos_indigenas']
negros_por_partido_ano = df.groupby(['ANO_ELEICAO', 'PARTIDO'])['Votos_Negros'].sum().reset_index()

# Filtrando para o ano de 2016 e ordenando
x = negros_por_partido_ano[negros_por_partido_ano['ANO_ELEICAO'] == 2024].sort_values('Votos_Negros')

x
```

```python
# 3. NM_VOTAVEL com maior número de votos por renda
# 4. NR_VOTAVEL com maior número de votos por renda
# Ordenar por renda para pegar as rendas mais altas e baixas
nm_votavel_renda = df.groupby(['ANO_ELEICAO','PARTIDO', 'RENDA' ])['QT_VOTOS'].sum().reset_index()
renda_sorted = nm_votavel_renda.sort_values(by='RENDA')


# Filtrando para o ano de 2016 e ordenando
x = renda_sorted[renda_sorted['ANO_ELEICAO'] == 2020].sort_values('RENDA')

x
```
